## Bin Statistics

```python
def percentile90(x):
    return(np.percentile(x, 90))

def bin_stats(df, relationship, stats, var_glm=None):
    '''calculate the statistic in each bin'''
    if relationship == 'cth-tbb':
        # median value
        cth_bins = np.arange(0, 20.2, 0.2),
        tbb_bins = np.arange(170, 301, 1)
        if stats == 'median':
            ret = binned_statistic_2d(df['HT'],
                                      df['C13'],
                                      df[var_glm],
                                      statistic=np.median,
                                      bins=[cth_bins, tbb_bins])
        elif stats == 'percentile90':
            ret = binned_statistic_2d(df['HT'],
                                      df['C13'],
                                      statistic=percentile90,
                                      bins=[cth_bins, tbb_bins])
        return ret.statistic, ret.x_edge, ret.y_edge

    elif relationship == 'fed-cth':
        fed_bins = np.arange(0, 81, 1)
        cth_bins = np.arange(0, 20.2, 0.2)
        if stats == 'frequency':
            ret = binned_statistic_2d(df['flash_extent_density'],
                                      df['HT'],
                                      values=None,
                                      statistic='count',
                                      bins=[fed_bins, cth_bins])
        elif stats == 'p90':
            df = df.loc[(df['HT'] >0) & (df['HT'] <20) & 
                        (df['flash_extent_density']>0) & (df['flash_extent_density']<80)]
            ret = binned_statistic(df['HT'],
                                   df['flash_extent_density'],
                                   statistic=percentile90,
                                   bins=[cth_bins])
    elif relationship == 'w-cth':
        # frequency
        w_bins = np.arange(0, 30.5, 0.5)
        cth_bins = np.arange(0, 20.2, 0.2)
        if stats == 'frequency':
            ret = binned_statistic_2d(df['w'],
                                      df['HT'],
                                      values=None,
                                      statistic='count',
                                      bins=[w_bins, cth_bins])
        elif stats == 'p90':
            df = df.loc[(df['HT'] >0) & (df['HT'] <20) & (df['w']>0) & (df['w']<30)]
            ret = binned_statistic(df['HT'],
                                   df['w'],
                                   statistic=percentile90,
                                   bins=[cth_bins])
    if stats == 'frequency':
        return ret.statistic/ret.statistic.sum(), ret.x_edge, ret.y_edge
    elif stats == 'p90':
        return ret.statistic, ret.bin_edges, ret.binnumber

def plot_stats(ax, x, y, stats, colorbar_label,
               vmin=None, vmax=None,
               cmap='viridis', cmap_kw=None,
               plot_clb=False,
               title=None):
    if not vmax:
        vmax = np.nanpercentile(stats, 95)
    m = ax.pcolormesh(x, y, stats,
                      vmin=vmin,
                      vmax=vmax,
                      cmap=cmap,
                      cmap_kw=cmap_kw)
    if plot_clb:
        ax.colorbar([m], label=colorbar_label)
    ax.format(title=title)
```
